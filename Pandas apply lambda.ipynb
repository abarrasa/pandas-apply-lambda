```python
dataset_votes = dataset['Votes']
dataset_votes
```

```python
def create_bins(dataset_votes):
    if dataset_votes >= 0 and dataset_votes <= 999:
        return 1
    elif dataset_votes >= 1000 and dataset_votes <= 9999:
        return 2
    elif dataset_votes >= 10000 and dataset_votes <= 99999:
        return 3
    elif dataset_votes >= 100000 and dataset_votes <= 999999:
        return 4
    elif dataset_votes >= 1000000:
        return 5 
    else:
        return "Error"
```

```python
bin = dataset["Votes"].apply(create_bins)
bin #hemos creado una serie pero no la hemos includio en el dataframe
```

```python
dataset["bin"] = dataset["Votes"].apply(create_bins)
```

## Challenge 3

We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

```python
def new_ranking(x,y):
    if 'Thriller' in x:
        return y +1
    elif 'Comedy' in x:
        return y -1
    else: 
        return y
```

```python

dataset['Rating'] = dataset.apply(lambda x: new_ranking(x['Genre'], x['Rating']), axis=1)
dataset
```
